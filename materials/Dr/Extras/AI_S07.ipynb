Import tensor flow and create a constant tensor of shape and value your choice, then print it

```python
import tensorflow as tf
t = tf.constant([3,4,6,7], shape = [2,2], dtype=tf.float32)
print(t)
```

Print the value of the tensor only

```python
print(t.numpy())
```

Create a ones tensor like the constant tensor you just created

```python
t1 = tf.ones_like(t)
print(t1)
```

Create a variable tensor using the ones tensor you just created, and read its values

```python
tv = tf.Variable(t1)
tv.read_value()
print(tv.numpy())
```

Add ones to the variable tensor and print it

```python
tadd=tv.assign_add(tf.ones([2,2]))
print(tadd.numpy())
```

Create 4 images 100X100X3 using random tensor normal dist. 
Import pyplot, then access the first image only and draw it

```python
img_t = tf.random.normal([4,100,100,3])
import matplotlib.pyplot as plt
plt.imshow(img_t[0,:,:,:])#plt.imshow(img_t[0,...])
```

Display only a slice of the image (from 40 to 70 in the height and from 40 to 60 in the width) (as if we are zooming in)

```python
plt.imshow(img_t[0,40:70, 40:60, :])
```

Create an image tensor of size 50X100X3, display it and then reshape it to 100X50X3, then display (draw) the reshaped image

```python
img_t2 = tf.random.normal([50,100,3])
plt.imshow(img_t2)
```

```python
img_reshape = tf.reshape(img_t2, shape=[ 100, 50,3])
plt.imshow(img_reshape)
```

Create an image then make it brighter (add a tensor of constant value)


```python
t_img1 = tf.random.normal([100,100,3])
plt.imshow(t_img1)
t_bright_factor = tf.fill([100,100,3], 0.6)# try with greater values too
plt.imshow(t_bright_factor)
tbright = t_img1+t_bright_factor
plt.imshow(tbright)
```
